##import Dataset

```python
dataset = pd.read_csv('data.tsv', sep='\t')
```

```python
arr = np.array(dataset)
inArr = arr
outArr = arr
```

##Input/Output separation 

```python
inputSet = inArr[:, [0,1,2,3,4,5,6,7]]
outputSet = outArr[:, [8]]
```

##Train Test separation

```python
trainData, testData, trainOut, testOut = train_test_split(inputSet, outputSet, test_size=0.2, random_state=12)
```

##Implementing Nodes and layers

```python
model = keras.Sequential(
    [
        keras.layers.Dense(units=1024, input_dim=8),
        keras.layers.Dense(units=512),
        keras.layers.Dense(units=256),
        keras.layers.Dense(units=128),
        keras.layers.Dense(units=64),
        keras.layers.Dense(units=32),
        keras.layers.Dense(units=16),
        keras.layers.Dense(units=8),
        keras.layers.Dense(units=4),
        keras.layers.Dense(units=2),
        keras.layers.Dense(units=1, activation=tf.keras.activations.sigmoid, name='output_layer')
    ]
)
```

```python
# model.compile(optimizer = tf.keras.optimizers.Adam(), loss = "binary_crossentropy", metrics = [tf.keras.metrics.AUC()])
model.compile(optimizer = tf.keras.optimizers.Adam(), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

##Train the Neural Network

```python
model.fit(trainData, trainOut, validation_data=(testData, testOut), epochs=10000, verbose=1)
```

##Predicting Output / Accuracy Measurement

```python
y_hat = model.predict(testData)

y = testOut.reshape(-1)
y_hat = np.round(y_hat.reshape(-1))

print('actual Output:\n', y)
print('Predicted Output:\n', y_hat)

print('Accuracy Score: ',np.round(accuracy_score(y, y_hat)*100 , 2), '%')
```
